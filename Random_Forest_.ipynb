```python
titanic=pd.read_csv('/content/titanic_data.csv',usecols=[1,2,4,11])
```

```python
titanic
```

```python
titanic=titanic.interpolate()
print(titanic)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()


sex=label.fit_transform(titanic['sex'])
Embarked=label.fit_transform(titanic['embarked'])

titanic['sex']=sex
titanic['embarked']=embarked
print(titanic)
```

```python
X=titanic.drop(['Survived'],1)
y=titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=.75,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_1=RandomForestClassifier(n_estimators=100)
model_1.fit(xtrain,ytrain)
```

```python
model_1.score(xtest,ytest)
```
